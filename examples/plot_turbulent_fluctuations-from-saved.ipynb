```python
import cmasher as cmr
%matplotlib widget
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm, Normalize
```

```python
def plot_density():
    dens_tot = slicer.slice_variable('0_Density')
    dens_bulk = slicer.slice_variable(smoothVar)
    dens_turb = slicer.slice_variable(relativeTurbVar)
    
    dens_tot = dens_tot.to_physical.cgs
    dens_bulk = dens_bulk.to_physical.cgs
    # dens_turb = dens_turb.to_physical.cgs
    
    vmin = np.min(dens_tot.value)
    vmax = np.max(dens_tot.value)
    norm = LogNorm(vmin=vmin,vmax=vmax)
    
    
    vmin = -0.8*dens_turb.max()
    vmax = 0.8*dens_turb.max()
    norm1 = Normalize(vmin=vmin,vmax=vmax)
    cmap = cmr.fusion
    
    fig, axes = plt.subplots(1,3, figsize=(12,6),sharex=True, sharey=True)
    
    im1 = axes[0].imshow(
        dens_tot.value, origin='lower', norm=norm,extent=extent.value)
    im2 = axes[1].imshow(
        dens_bulk.value, origin='lower', norm=norm,extent=extent.value)
    im3 = axes[2].imshow(
        dens_turb, origin='lower', norm=norm1,cmap=cmap,extent=extent.value)
    
    axes[0].set_title('Total')
    axes[1].set_title('Bulk')
    axes[2].set_title('Turbulent')
    
    for i in range(3):
        axes[i].set_xlabel(extent.label('x'))
    
    axes[0].set_ylabel(extent.label('y'))
    
    
    
    cbar = fig.colorbar(im1, orientation='horizontal')
    cbar.set_label(dens_tot.label('\\rho_0'))
    
    cbar = fig.colorbar(im2, orientation='horizontal')
    cbar.set_label(dens_tot.label('\\rho_{\mathrm{smooth}}'))
    
    cbar = fig.colorbar(im3, orientation='horizontal')
    cbar.set_label(r'$\delta \rho / \rho_0 $')
    
    
    
    fig.subplots_adjust(top=0.9,
    bottom=0.1,
    left=0.10,
    right=0.90,
    hspace=0.0,
    wspace=0.05)
    
    fig.suptitle('Density field (iterative filter)',y=0.95)
    plt.savefig(f'Halo{halo}_density_center_non_starforming.pdf')
    plt.close()
```

```python
def plot_kinetic_energy():
    snap['0_KineticEnergy'] = 0.5*snap['0_Density'] * (snap['0_Velocities'][:,0]**2 + snap['0_Velocities'][:,1]**2 +
                                   snap['0_Velocities'][:,2]**2)
    
    kin_en_tot = slicer.slice_variable('0_KineticEnergy')
    kin_en_bulk = slicer.slice_variable(smoothEner)
    kin_en_turb = slicer.slice_variable(turbEner)
    
    kin_en_tot = kin_en_tot.to_physical.cgs
    kin_en_bulk = kin_en_bulk.to_physical.cgs
    kin_en_turb = kin_en_turb.to_physical.cgs
    
    vmin = 0.01*kin_en_tot.value.max()
    vmax = 0.8*kin_en_tot.value.max()
    norm = LogNorm(vmin=vmin,vmax=vmax)
    
    
    vmin = 0.01*kin_en_turb.value.max()
    vmax = 0.8*kin_en_turb.value.max()
    norm1 = LogNorm(vmin=vmin,vmax=vmax)
    # cmap = cmr.fusion
    
    fig, axes = plt.subplots(1,3, figsize=(12,6),sharex=True, sharey=True)
    
    im1 = axes[0].imshow(
        kin_en_tot.value, origin='lower', norm=norm,extent=extent.value)
    im2 = axes[1].imshow(
        kin_en_bulk.value, origin='lower', norm=norm,extent=extent.value)
    im3 = axes[2].imshow(
        kin_en_turb.value, origin='lower', norm=norm1,extent=extent.value)
    
    axes[0].set_title('Total')
    axes[1].set_title('Bulk')
    axes[2].set_title('Turbulent')
    
    for i in range(3):
        axes[i].set_xlabel(extent.label('x'))
    
    axes[0].set_ylabel(extent.label('y'))
    
    
    
    cbar = fig.colorbar(im1, orientation='horizontal')
    cbar.set_label(kin_en_tot.label('1/2 \, \\rho v^2'))
    
    cbar = fig.colorbar(im2, orientation='horizontal')
    cbar.set_label(kin_en_tot.label('1/2 \, \\rho v_{\mathrm{smooth}}^2'))
    
    cbar = fig.colorbar(im3, orientation='horizontal')
    cbar.set_label(kin_en_tot.label('1/2 \, \\delta \\rho v^2'))
    
    fig.subplots_adjust(top=0.9,
    bottom=0.1,
    left=0.10,
    right=0.90,
    hspace=0.0,
    wspace=0.05)
    
    fig.suptitle('Kinetic energy (iterative filter)',y=0.95)
    plt.savefig(f'Halo{halo}_kin_energy_center_non_starforming.pdf')
    plt.close()
```

```python
def plot_magnetic_energy():
    snap['0_MagneticEnergy'] = 0.5*(snap['0_MagneticField'][:,0]**2 + snap['0_MagneticField'][:,1]**2 +
                                   snap['0_MagneticField'][:,2]**2)
    
    mag_en_tot = slicer.slice_variable('0_MagneticEnergy')
    mag_en_bulk = slicer.slice_variable(smoothMagEner)
    mag_en_turb = slicer.slice_variable(turbMagEner)
    
    mag_en_tot = mag_en_tot.to_physical.cgs
    mag_en_bulk = mag_en_bulk.to_physical.cgs
    mag_en_turb = mag_en_turb.to_physical.cgs
    
    vmin = 0.01*mag_en_tot.value.max()
    vmax = 0.8*mag_en_tot.value.max()
    norm = LogNorm(vmin=vmin,vmax=vmax)
    
    
    vmin = 0.01*mag_en_turb.value.max()
    vmax = 0.8*mag_en_turb.value.max()
    norm1 = LogNorm(vmin=vmin,vmax=vmax)
    # cmap = cmr.fusion
    
    fig, axes = plt.subplots(1,3, figsize=(12,6),sharex=True, sharey=True)
    
    im1 = axes[0].imshow(
        mag_en_tot.value, origin='lower', norm=norm,extent=extent.value)
    im2 = axes[1].imshow(
        mag_en_bulk.value, origin='lower', norm=norm,extent=extent.value)
    im3 = axes[2].imshow(
        mag_en_turb.value, origin='lower', norm=norm1,extent=extent.value)
    
    axes[0].set_title('Total')
    axes[1].set_title('Bulk')
    axes[2].set_title('Turbulent')
    
    for i in range(3):
        axes[i].set_xlabel(extent.label('x'))
    
    axes[0].set_ylabel(extent.label('y'))
    
    
    
    cbar = fig.colorbar(im1, orientation='horizontal')
    cbar.set_label(mag_en_tot.label('1/2 \, B^2'))
    
    cbar = fig.colorbar(im2, orientation='horizontal')
    cbar.set_label(mag_en_tot.label('1/2 \, B_{\mathrm{smooth}}^2'))
    
    cbar = fig.colorbar(im3, orientation='horizontal')
    cbar.set_label(mag_en_tot.label('1/2 \, \\delta B^2'))
    
    fig.subplots_adjust(top=0.9,
    bottom=0.1,
    left=0.10,
    right=0.90,
    hspace=0.0,
    wspace=0.05)
    
    fig.suptitle('Magnetic energy (iterative filter)',y=0.95)
    plt.savefig(f'Halo{halo}_mag_energy_center_non_starforming.pdf')
    plt.close()
```

```python
# halo = halonums[0]
for halo in halonums:
    print(f'Making plots for halo: {halo}')
    simfolder = pico.get_simfolder(halo)
    info = pico.PicoOutputInfo(f'{simfolder}output/pico_output_info.txt')
    tracking = np.loadtxt(f'{simfolder}output/pico_tracking_info.txt', dtype=int)
    
    snapnum = info.num_vec_full[::-1][0]
    
    analysis_folder = pico.get_analysisfolder(halo) + 'turbulence/'
    
    turb = pa.PaicosReader(analysis_folder, snapnum, basename='filtered')
    
    snap = pa.Snapshot(simfolder + 'output/', snapnum, basename='snapshot')
    
    index = ~(snap['0_StarFormationRate'] > 0.)
    snap = snap.select(index, parttype=0)
    
    center = snap.Cat.Group['GroupPos'][0]
    R200 = snap.Cat.Group['Group_R_Crit200'][0]
    widths = 0.5 * np.ones(3) * R200
    
    widths_slicer = widths.copy
    widths_slicer[2] = 0.
    slicer = pa.Slicer(snap, center, widths_slicer, 'z', npix=2048)
    extent = slicer.centered_extent.to('Mpc')
    
    sf_index = turb['sf_index']
    turbVar = turb['TurbDensity']
    smoothVar = turb['smoothDensity']
    smoothEner = turb['smoothEner']
    turbEner = turb['turbEner']
    smoothMagEner = turb['smoothMagEner']
    turbMagEner = turb['turbMagEner']
    
    relativeTurbVar = np.zeros(snap['0_Density'].shape)
    relativeTurbVar[sf_index] = turbVar[sf_index].value/snap['0_Density'][sf_index].value
    
    plot_density()
    plot_kinetic_energy()
    plot_magnetic_energy()
```
